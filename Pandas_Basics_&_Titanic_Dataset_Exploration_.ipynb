```python
#Day 3: Pandas Basics & Titanic Dataset Exploration

# 1. Εισαγωγή βιβλιοθηκών
import pandas as pd
import matplotlib.pyplot as plt

# 2. Φόρτωμα dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# 3. Ματιά στα δεδομένα
print("----- 5 πρώτες γραμμές -----")
print(df.head())

print("\n----- Πληροφορίες DataFrame -----")
print(df.info())

print("\n----- Στατιστικά -----")
print(df.describe())

# 4. Επιλογές δεδομένων
print("\nΌνομα πρώτου επιβάτη:", df.loc[0, "Name"])
print("Ηλικία πρώτου επιβάτη:", df.loc[0, "Age"])

# 5. Mini Ερωτήσεις
print("\n--- Mini Analysis ---")
print("Σύνολο επιβατών:", len(df))
print("Επιζώντες:", df["Survived"].sum())
print("Ποσοστό επιζώντων:", df["Survived"].mean() * 100, "%")
print("Μέση ηλικία:", df["Age"].mean())
print("Επιβάτες ανά φύλο:")
print(df["Sex"].value_counts())

# 6. Οπτικοποίηση: κατανομή ηλικίας
plt.hist(df["Age"].dropna(), bins=20, edgecolor="black")
plt.title("Κατανομή Ηλικιών Επιβατών")
plt.xlabel("Ηλικία")
plt.ylabel("Αριθμός Επιβατών")
plt.show()

# 7. Οπτικοποίηση: επιζώντες ανά φύλο
df.groupby("Sex")["Survived"].mean().plot(kind="bar", color=["blue", "pink"])
plt.title("Ποσοστό Επιζώντων ανά Φύλο")
plt.ylabel("Ποσοστό Επιβίωσης")
plt.show()
```
